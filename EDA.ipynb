# Load files

```python
# Standard imports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# OS
import os

# Set the option to display all columns
pd.set_option('display.max_columns', None)
```

```python
# Use Data directory
FOLDER = "Data/"
os.makedirs(FOLDER, exist_ok=True)
os.listdir(FOLDER)
```

```python
# Load the JSON files into dataframes
df_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
df_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')

df_2000.head()
```

```python
# Concatenate the dataframes into a single dataframe
combined_df = pd.concat([df_2000, df_2001], ignore_index=True)
combined_df
```

```python
combined_df.info()
```

```python
percentage_missing = combined_df.isna().mean() * 100
percentage_missing
```

# EDA

How many movies had at least some valid financial information (values > 0 for budget OR revenue)?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Boolean mask for filtering movies with valid financial information
valid_financial_info_mask = (combined_df['budget'] > 0) | (combined_df['revenue'] > 0)

# Use the mask to filter the dataframe and count the movies
valid_financial_movies = combined_df[valid_financial_info_mask]

# Exclude movies with 0 values for both budget and revenue
valid_nonzero_movies = valid_financial_movies[(valid_financial_movies['budget'] > 0) | (valid_financial_movies['revenue'] > 0)]

# Create countplot
plt.figure(figsize=(8, 6))
sns.countplot(data=valid_nonzero_movies, x='certification')
plt.xlabel('Certification')
plt.ylabel('Count of Movies')
plt.title("Count of Movies with Valid Financial Information (Excluding 0's for Budget and Revenue)")
plt.show()

# Get the total count of movies with valid financial information
total_valid_nonzero_movies = valid_nonzero_movies.shape[0]
print(f'Total Movies with Valid Financial Information (Excluding 0s for Budget and Revenue): {total_valid_nonzero_movies}')
```

How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter the DataFrame for movies with valid financial information and non-zero budget/revenue
valid_nonzero_movies = combined_df[(combined_df['budget'] > 0) | (combined_df['revenue'] > 0)]

# Create a countplot to count movies in each certification category
plt.figure(figsize=(8, 6))
sns.countplot(data=valid_nonzero_movies, x='certification', order=['G', 'PG', 'PG-13', 'R'])
plt.xlabel('Certification')
plt.ylabel('Count of Movies')
plt.title('Count of Movies in Each Certification Category')
plt.show()
```

What is the average revenue per certification category?

```python
# Filter the DataFrame for movies with valid financial information and non-zero budget/revenue
valid_nonzero_movies = combined_df[(combined_df['budget'] > 0) | (combined_df['revenue'] > 0)]

# Calculate the average revenue per certification category
average_revenue_by_certification = valid_nonzero_movies.groupby('certification')['revenue'].mean().reset_index()

# Create a barplot to visualize the average revenue per certification category
plt.figure(figsize=(8, 6))
sns.barplot(data=average_revenue_by_certification, x='certification', y='revenue', order=['G', 'PG', 'PG-13', 'R'])
plt.xlabel('Certification')
plt.ylabel('Average Revenue')
plt.title('Average Revenue per Certification Category')
plt.xticks(rotation=0)  # Rotate x-axis labels for readability

# Format the y-axis ticks with commas for readability
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))

plt.show()
```

What is the average budget per certification category?


```python
# Filter the DataFrame for movies with valid financial information and non-zero budget/revenue
valid_nonzero_movies = combined_df[(combined_df['budget'] > 0) | (combined_df['revenue'] > 0)]

# Calculate the average budget per certification category
average_budget_by_certification = valid_nonzero_movies.groupby('certification')['budget'].mean().reset_index()

# Create a barplot to visualize the average budget per certification category
plt.figure(figsize=(8, 6))
sns.barplot(data=average_budget_by_certification, x='certification', y='budget', order=['G', 'PG', 'PG-13', 'R'])
plt.xlabel('Certification')
plt.ylabel('Average Budget')
plt.title('Average Budget per Certification Category')

# Format the y-axis ticks with commas for readability
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:,.0f}'.format(x)))

plt.show()
```
