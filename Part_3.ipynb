## Title_Basics Table

```python
# Check title_basics column names for accuracy in normalization steps
title_basics_df.head()
```

```python
# Create a column with list of genres
title_basics_df['genres_split'] = title_basics_df['genres'].str.split(',')
title_basics_df
```

```python
# Explode the list of genres into new rows
exploded_genres = title_basics_df.explode('genres_split')
exploded_genres
```

```python
# Sort genres alphabetically
unique_genres = sorted(exploded_genres['genres_split'].unique())
```

```python
# Create title_genres table
title_genres_df = exploded_genres[['tconst', 'genres_split']].copy()
title_genres_df.head()
```

```python
# Map genres in dictionary
genre_ints = range(len(unique_genres))
genre_map = dict(zip(unique_genres, genre_ints))
genre_map
```

```python
# Make new integer column in title_genres table
title_genres_df['genre_id'] = title_genres_df['genres_split'].map(genre_map)
title_genres_df = title_genres_df.drop(columns = 'genres_split')
title_genres_df.head()
```

```python
# Convert genre map dictionary into a table
genres_df = pd.DataFrame({'Genre_Name': genre_map.keys(),
                      'Genre_ID': genre_map.values()})
genres_df.head()
```

```python
# Before dropping columns
display(title_basics_df.info())

# Drop columns from title_basics df
title_basics = title_basics_df.drop(columns = ['isAdult', 'titleType', 
                                            'originalTitle', 'endYear', 'genres', 
                                            'genres_split'])

# After dropping columns
display(title_basics.info())
```

# Add Tables to Database

### Functions

```python
def query_top_5(table_name):
    # Query to run first five rows of a table
    q = f"""
    SELECT *
    FROM {table_name}
    LIMIT 5;
    """
    return pd.read_sql(q, engine)
```

### Title Basics

```python
# Check dtypes
title_basics.dtypes
```

```python
from sqlalchemy.types import *

# Calculate max string lengths for object columns
id_len = title_basics['tconst'].fillna('').map(len).max()
title_len = title_basics['primaryTitle'].fillna('').map(len).max()

# Create schema dict using sqlalchemy datatype objects
basics_schema = {
    "tconst": String(id_len+1), 
    "primaryTitle": Text(title_len+1),
    'startYear':Float(),
    'runtimeMinutes':Integer()}
```

```python
# Save table to sql
title_basics.to_sql('title_basics', engine, dtype = basics_schema,
                    if_exists = 'replace', index = False)
```

```python
# Add tconst as primary key
engine.execute('ALTER TABLE title_basics ADD PRIMARY KEY (`tconst`);')
```

```python
query_top_5('title_basics')
```
